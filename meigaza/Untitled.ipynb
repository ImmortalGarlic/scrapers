```python
import requests
from bs4 import BeautifulSoup

url_lst = ['http://miwaku-meigaza.com/joho.htm', 'http://miwaku-meigaza.com/joho2.htm', 'http://miwaku-meigaza.com/joho3.htm']

def get_schedule(url):
    response = requests.get(url)
    response.encoding = 'utf-8'
    soup = BeautifulSoup(response.text, 'html.parser')
    section = soup.find_all('section')
    for s in section:
        cinema = s.find('h2')
        films = s.find_all('p')
        if cinema is None:
            continue
        cinema = cinema.text
        films = [x.text for x in films]
        print (films)
```

```python
get_schedule(url_lst[0])
```
